```python
# Drop the unused column
# Process 1
# dataset.drop(['RowNumber','CustomerId','Surname'], axis = 1, inplace = True)
# Process 2
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

## Encoding Categorical Data

### Label Encoding the "gender' column

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
```

```python
print(X)
```

### OneHot Encoding Geography column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[1])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

## Splitting the dataset into the training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .20, random_state = 0)
```

```python
print('X_train shape',X_train.shape)
print('X_test shape', X_test.shape)
print('y_train shpae', y_train.shape)
print('y_test shape', y_test.shape)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## Intializing the ANN

```python
ann = tf.keras.models.Sequential()
```

### Adding the input layer and first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6,activation= 'relu'))
```

### Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6 , activation='relu'))
```

### Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid' ))
```

## Training the output layer

### Compiling the ANN

```python
ann.compile(optimizer = 'adam', loss= 'binary_crossentropy' , metrics= ['accuracy'])
```

### Training the ANN on the Training set

```python
ann.fit(X_train, y_train, batch_size= 32, epochs = 100)
```

## Part 4: making the prediction and evaluating the model

Therefore, Our ANN model predicts that this customer stays in the bank

Import Note1: Notice that the values of the features were all input in a double pair of square brackets. That's beacuse the 'predict' method always expect the 2D array as the format of its inputs & putting our values into a double pair of square brackets makes the input exactly a 2D array

Important Note 2:
Notices also that the "France" country was not impute as a string in the last column but as "1,0,0" in the first three column. That's because of course the predict method expect the One-Hot encoded values of the state, & as we see in the first row of the matix of features X,"France" was encoded as "1,0,0". and be careful to include these values in the first three columns, because the dummy varaibles are always created in the first column.

## Predict the test result

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > .5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
